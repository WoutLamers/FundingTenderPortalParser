```python
import requests
import json
from pprint import pp
```

```python
url = 'https://api.tech.ec.europa.eu/search-api/prod/rest/facet'

params = dict(
    text='***',
    apiKey='SEDIA',
)
resp = requests.post(url=url, params=params)
facets = resp.json()
pp(resp.status_code)
```

```python
pp({d:('list len '+str(len(v)) if isinstance(v,list) else v) for d, v in facets.items()})
pp('Each list entry of "facets" is a dict with "name" and list of "values"')
pp({f['name']: len(f['values']) for f in facets.get('facets')})
```

```python
facetname = 'destination'
{v['rawValue']: v['value'] for v in [f for f in facets['facets'] if f.get('name') == facetname][0]['values']}
```

```python
frameworkProgrammes = {v['rawValue']: v['value'] for v in [f for f in facets['facets'] if f.get('name') == "frameworkProgramme"][0]['values']}
pp(frameworkProgrammes)
```

```python
callIdentifiers = [f['values'] for f in facets.get('facets') if f['name']=="callIdentifier"][0]
pp(len([(v['rawValue'], v['value']) for v in callIdentifiers if v['rawValue']!=v['value']]))
callIdentifiers = [v['value'] for v in callIdentifiers]
pp(callIdentifiers[:6])
```

```python
"We can search for Calls in general"

url = 'https://api.tech.ec.europa.eu/search-api/prod/rest/search'

params = dict(
    text='***',
    apiKey='SEDIA',
    pageSize='100',
)
query={
  "bool": {
    "must": [
      {
        "terms": {
          "type": [
            "1",
            "2",
            "8"
          ]
        }
      },
      {
        "terms": {
          "status": [
            "31094501",
            "31094502",
            "31094503"
          ]
        }
      },
      {
        "term": {
          "programmePeriod": "2021 - 2027"
        }
      },
      {
        "terms": {
          "frameworkProgramme": [
            "43108390"
          ]
        }
      },
      {
        "terms": {
          "crossCuttingPriorities": [
            "PART-COPROGR"
          ]
        }
      },
      {
        "terms": {
          "destination": [
            "43650651"
          ]
        }
      },
      # {
      #   "term": {
      #     "callIdentifier": "HORIZON-INFRA-2021-EOSC-01"
      #   }
      # }
    ]
  }
}

languages = ["en"]
sort = {"field": "sortStatus", "order": "ASC"}

resp = requests.post(url=url, params=params, files={
    'query': ('blob', json.dumps(query), 'application/json'),
    'languages': ("blob", json.dumps(languages), "application/json"),
    'sort': ("blob", json.dumps(sort), "application/json")
 }
)
data = resp.json()
pp(data)
```
